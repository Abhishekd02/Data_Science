# Line plot

plt.plot()

```python
ypoints = np.array([3,8,1,10,16,5])
plt.plot(ypoints)  #plot()-->line plot
plt.xlabel("Count of ypoints")  #name of x axis
plt.ylabel("Points of y")  #name of y axis
plt.title("Line plot")  #title for whole graph
plt.show()  #display the graph
```

```python
speed = [10,20,30]
cal1 = [100,127,225]
plt.plot(speed,cal1,marker="*")
plt.xlabel("speed")
plt.ylabel("calories burnt")
plt.title("Line")
plt.show()
```

```python
speed = [10,20,30]
cal1 = [100,127,225]
plt.plot(speed,cal1,marker="*",linestyle="dotted")
plt.xlabel("speed")
plt.ylabel("calories burnt")
plt.title("Line")
plt.show()
```

```python
speed = [10,20,30]
cal1 = [100,127,225]
plt.plot(speed,cal1,marker="*",linestyle="dashed",color="red",markersize=20)
plt.xlabel("speed")
plt.ylabel("calories burnt")
plt.title("Line")
plt.show()
```

```python
speed = [10,20,30]
cal1 = [100,127,225]
cal2 = [170,200,290]
plt.plot(speed,cal1,marker="P",color="r",linestyle="dotted",label="sunday")
plt.plot(speed,cal2,marker="*",color="b",linestyle="dashed",label="monday")
plt.xlabel("speed")
plt.ylabel("calories burnt")
plt.title("Line")
plt.grid(True)
plt.legend()
plt.show()
```

# Bar plot

plt.bar(x,y)

```python
speed = [10,20,30]
cal1 = [100,300,160]
plt.bar(speed,cal1,width=4,color=["red","blue","orange"],edgecolor="Black")
plt.show()
```

```python
speed = [10,20,30]
cal1 = [100,300,160]
plt.barh(speed,cal1,color=["red","blue","orange"],edgecolor="Black")
plt.show()
```

```python
import matplotlib.pyplot as pl
import numpy as np
pl.figure(figsize=(10,7))
info=['Gold','Silver','Bronze']
india=[80,70,70]
aus=[55,50,45]
can=[34,67,89]
x=np.arange(len(info))
pl.bar(info,india,width=.15)
pl.bar(x+0.15,aus,width=.15)
pl.bar(x+0.30,can,width=.15)
pl.show()
```
